```python
def checking(data):
    l=[]
    columns=data.columns
    for col in columns:
        dtypes=data[col].dtypes
        nunique=data[col].nunique()
        sum_null=data[col].isnull().sum()
        l.append([col,dtypes,nunique,sum_null])
    new_df=pd.DataFrame(l)
    new_df.columns=['COLUMNS','DATA_TYPES','UNIQUE_VALUE','NULL?']
    return new_df
```

```python
checking(data)
```
